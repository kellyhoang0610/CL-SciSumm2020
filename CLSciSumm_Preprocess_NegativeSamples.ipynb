```python
import pandas as pd
import nltk

data_2018_file = "/efs/CL-SciSumm/Preprocessed_data/all_training_annotation_2018_cleaned_positive_negative.csv"
data_2018_df = pd.read_csv(data_2018_file)
print (len(data_2018_df))

#remove reference text that are Nan
data_2018_df = data_2018_df.dropna(subset=['reference_text'])
print (len(data_2018_df))

# remove reference text with too less words
for index, data in data_2018_df.iterrows():
    if len(nltk.word_tokenize(data["reference_text"])) <= 3:
        print (data["reference_text"])
        data_2018_df = data_2018_df.drop(index, inplace=False)

print (len(data_2018_df))

# data_2018_df.to_csv("/efs/CL-SciSumm/Preprocessed_data/all_training_annotation_2018_cleaned_positive_negative_filtered.csv")
```

```python
citing_text = (data_2018_df["citing_text"])
print (len(citing_text))
citing_text = set(citing_text)
print (len(citing_text))

citing_text_list = []
citing_text_positive_samples_count = []
citing_text_negative_samples_count = []

for text in citing_text:
    citing_text_list.append(text)
    count_positive = 0
    count_negative = 0
#     print (text)
    for index, data in data_2018_df.iterrows():
        if text == data["citing_text"] and data["label"]=="positive":
            count_positive += 1 
        if text == data["citing_text"] and data["label"]=="negative":
            count_negative += 1 
#     print (count_positive)
#     print (count_negative)
    citing_text_positive_samples_count.append(count_positive)
    citing_text_negative_samples_count.append(count_negative)
            
statistic_df = pd.DataFrame(list(zip(citing_text_list, citing_text_positive_samples_count,citing_text_negative_samples_count)), 
                                 columns = ["citance","positive_count","negative_count"]) 

statistic_df.to_csv("/efs/CL-SciSumm/Preprocessed_data/statistic_v1.csv")
```
